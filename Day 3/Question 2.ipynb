```python
# Using for loop please print all the prime numbers between 1- 200 using FOR LOOP AND RANGE function

for i in range(1, 200 + 1):
   
    if i > 1:
    
        for j in range(2, i):
        
            if (i % j) == 0:
                
                break
        else:
    
            print(i,end=", ")
    
```
