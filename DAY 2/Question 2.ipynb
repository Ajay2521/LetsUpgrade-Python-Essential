## Dictionary and its default functions.

```python
dict = {'Name': 'AJAY M', 'Age': 19}
```

```python
dict
```

### values()

```python
# values() used to gets values from the dict

dict.values()
```

### keys()

```python
# keys() used to gets keys from the dict

dict.keys()
```

### items()

```python
# items() used to get dict in tuple form

dict.items()
```

### get()

```python
# get() used to get specific value for given key

dict.get("Name")
```

### copy()

```python
# copy() one dict to another

dictCopy = dict.copy()
```

```python
dictCopy
```

### fromkeys()

```python
# fromkeys() used to create new dict by using keys only and values be default as none

newDict = dict.fromkeys(dict)
```

```python
newDict
```

### clear()

```python
# clear() used to clear the dict

dict.clear()
```
