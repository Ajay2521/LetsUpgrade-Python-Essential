## List and its default functions


```python
lst = [ 1, 2, 3, 4, 5 ]
```

```python
lst
```

### append()

```python
# append() to add element at the end of the list

lst.append(6)
```

```python
lst
```

### extend()

```python
# extend() used to add two or more list togther

lst.extend([ 7, 8, 9 ])
```

```python
lst
```

### insert()

```python
# insert() used to insert a element at specific place in the list

lst.insert(0, 0) # insert zero at the index 0
```

```python
lst
```

### remove()

```python
# remove() used to remove the element from the list

lst.remove(0) # removing element zero
```

```python
lst
```

### pop()

```python
# pop() used to remove last element or specific element

lst.pop() # remove last element
```

```python
lst
```

### count()

```python
# count() used to count number of occurence of specific element

lst.append(1)
lst.append(1)
lst.append(1)

lst.count(1)
```

```python
lst
```

### sort()

```python
# sort() used to sort the list

lst.sort()
```

```python
lst
```

### reverse()

```python
# reverse() used to reverse the list

lst.reverse()
```

```python
lst
```

### copy()

```python
# copy() used to copy one list to another list

lstCopy = lst.copy()
```

```python
lstCopy
```

### index()

```python
# index() used to return index value of sepcific element

lst.index(5)
```

### len()

```python
# len() used to return length of the list 

len(lst)
```

### min()

```python
# min() used to return the min value in the list

min(lst)
```

### max()

```python
# max() used to return the max value in the list

max(lst) 
```
