## Sets and its default functions.

```python
Set = {1, 2, 3, 4, 5, 2, 3, 4, 1, 2, 4, 1}

Set1 = {1, 4, 6, 7, 1, 8, 2 }
```

```python
Set
```

```python
Set1
```

### add()

```python
# add() used to add element to the set

Set.add(6)
```

```python
Set
```

### intersection()

```python
# intersection() used to return intersecting values of two or more sets

Set.intersection(Set1)
```

### union()

```python
# union() used to return all values of two or more sets

Set.union(Set1)
```

### remove()

```python
# remove() used to remove specific element from set

Set.remove(3)
```

```python
Set
```

### issuperset()

```python
# issuperset() used to return true if one set is super set of another set

Set.issuperset(Set1)
```

### issubset()

```python
# issubset() used to return true if one set is sub set of other

Set.issubset(Set1)
```

### difference()

```python
# difference() used to return only the item which is not common

Set.difference(Set1)
```
